#### One hot Encoding with many features
 so we took mercedes benz data set and we can see there are lots of categorical features so what we should we do to handle it?
 First we can make a list of top 10 most frequent features and then use one hot encoding to it.

```python
df = pd.read_csv("test.csv", usecols=["X0","X1","X2","X3","X4","X5","X6"])
df.head()
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
lst_10 = df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10 = list(lst_10)
lst_10
```

```python
# for now we just consider categories for "X1"
import numpy as np

for categories in lst_10:
    df[categories]=np.where(df["X1"]==categories,1,0)
```

```python
df[lst_10]
```

```python
#### This is just to show what we did
lst_10.append("X1")
df[lst_10]
```
